# GDP  basic line chart /jw


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('/content/drive/MyDrive/302groupshare/whrdata/2022.csv')

# Sort the data by happiness score and select the top 10 countries
df = df.sort_values("Happiness score", ascending=False).head(10)

# Adjust decimal points in numerical data
df['Happiness score'] = df['Happiness score'].str.replace(',', '.').astype(float).round(3)
df['Explained by: GDP per capita'] = df['Explained by: GDP per capita'].str.replace(',', '.').astype(float).round(3)

# Create a line chart
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(df["Country"], df["Explained by: GDP per capita"], color="skyblue", alpha=0.6)
ax.plot(df["Country"], df["Happiness score"], color="navy", marker="o")

# Add data labels to the chart

for i, (score, gdp, country) in enumerate(zip(df["Happiness score"], df["Explained by: GDP per capita"], df["Country"])):
    ax.text(country, score, f'{score:.2f}', ha='center', va='bottom')
    ax.text(country, gdp, f'{gdp:.2f}', ha='center', va='bottom')

# Set the chart title, axis labels, and limits
ax.set_title("Top 10 Happiest Countries in 2022 vs GDP")
ax.set_xlabel("Country")
ax.set_ylabel("Happiness Score / Explained by: GDP per capita")
ax.set_ylim(0, 10)

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45, ha='right')

# Show the chart
plt.show()
```

# TOP 10 / Lollipop Bar Chart /jw

```python

import pandas as pd
import matplotlib.pyplot as plt

# Load the data from the CSV file
df = pd.read_csv('/content/drive/MyDrive/302groupshare/whrdata/2022.csv')

# Sort the data by happiness score and select the top 10 countries
df = df.sort_values("Happiness score", ascending=False).head(10)

# Adjust decimal points in numerical data
df['Happiness score'] = df['Happiness score'].str.replace(',', '.').astype(float).round(3)


# Create a lollipop chart
fig, ax = plt.subplots(figsize=(10, 6))

ax.hlines(y=df["Country"], xmin=0, xmax=df["Happiness score"], color="lightblue")
ax.plot(df["Happiness score"], df["Country"], "o", color="navy")

# Add data labels to the chart
for i, (score, country) in enumerate(zip(df["Happiness score"], df["Country"])):
    ax.text(score + 0.2, i, str(score), ha='left', va='center')

# Set the chart title and axis labels
ax.set_title("Countdown of Top 10 Happiest Countries in 2022")
ax.set_xlabel("Happiness Score")
ax.set_ylabel("Country")
ax.set_xlim(0, 10)
ax.set_ylim(-1, len(df))

# Show the chart
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Read the data
df = pd.read_csv('/content/drive/MyDrive/World Happiness Data.csv')

# Sort by 'Happiness Score' in descending order
df_sorted = df.sort_values('Happiness score', ascending=False)

# Select top 10 countries by 'Happiness Score'
top_10 = df_sorted.head(10)

# Reset index to start from 1
top_10.index = top_10.index + 1

# Select columns 'Country', 'Explained by: Healthy life expectancy', and 'Happiness Score'
subset = top_10[['Country', 'Explained by: Healthy life expectancy', 'Happiness score']]

# Rename the columns
subset = subset.rename(columns={'Explained by: Healthy life expectancy': 'Life Expectancy'})

# Adjust decimal points in numerical data
subset['Life Expectancy'] = subset['Life Expectancy'].str.replace(',', '.').astype(float).round(3)
subset['Happiness score'] = subset['Happiness score'].str.replace(',', '.').astype(float).round(3)

# Set the index name to blank
subset.index.name = ''

# Define a list of colors for each country
colors = plt.cm.tab20c.colors

# create a scatter plot
fig, ax = plt.subplots(figsize=(10, 6)) # Set the size of the figure
for i, row in subset.iterrows():
    ax.scatter(row['Happiness score'], row['Life Expectancy'], color=colors[i-1], label=row['Country'], s=150)

ax.set_xlabel('Happiness score')
ax.set_ylabel('Life Expectancy')
ax.set_title('Happiness Score vs. Life Expectancy by Country')

# Format the y-axis tick labels to display values as 750 instead of 0.75
plt.ticklabel_format(axis='y', style='plain')
ax.yaxis.set_major_formatter(plt.FuncFormatter(lambda x, loc: "{:,}".format(int(x*1000))))

ax.legend(loc='upper left', bbox_to_anchor=(1.05, 1)) # Show the legend outside the plot

# Add a grid to the plot
ax.grid()

# Set the background color of the plot
fig.set_facecolor('#f0f0f0')

plt.show()
```
